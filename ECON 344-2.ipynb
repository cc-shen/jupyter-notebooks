## Chapter 10: Developing New Products and Services

### The Variations of Products

**Product Line**

> A group of products that are closely related because they satisfy a class of need, are used together, are sold to same customer group, are distributed through the same type of outlets, and fall within a given price range.
>
> Each product line has its own marketing strategy

**Product Mix**

> The number of product lines offered by a company

Classifying products,

* type of user
    * consumer goods
    * business goods
* degree of tangibility

#### Classifying Consumer and Business Goods

##### Classification of Consumer Goods

**Convenience goods**

> Purchased frequently with minimum shopping effort, while consumers compare **shopping goods** on several attributes

**Specialty goods**

> Items for which the consumer makes a special effort to obtain

**Unsought goods**

> Products in which the consumer either does not know about or knows about but doe not initially want

##### Classification of Business Goods

**Production goods**

> Become part of the final product

**Support goods**

> Purchased to assist in the production of the final product

### What is a New Product

_New_ is difficult to define,

* newness compared with existing products
* newness in legal terms
* newness from the company's perspective
* newness from the consumer's perspective

Consumption effects define newness, i.e. degree of new consumer learning needed:

* High
    * Discontinuous Innovations
        * requires new learning and consumption patterns by consumers
    * Marketing strategy
        * educate consumers through product trial and personal selling
* Medium
    * Dynamically continuous innovations
        * disrupts consumer's normal routine but does not require totally new learning
    * Marketing strategy
        * advertise points of difference and benefits to consumers
* Low
    * Continuous innovations
        * requires no new learning by consumers
    * Marketing strategy
        * gain consumer awareness and wide distribution

### Why New Products or Services Succeed or Fail

Marketing reasons for new-product failures

* insignificant "point of difference"
* incomplete market and product definition
* too little market attractiveness
* poor execution of the marketing mix
* poor product quality on critical factors
* bad timing
* no economical access to buyers

A look at some failures

* What were they thinking? Organizational problems new-product failure
    * not really listening to the "voice of the consumer"
    * skipping steps in the new-product process
    * pushing a poorly conceived product into the market to generate quick revenue
    * "Groupthink" in task force and committee meetings
    * not learning critical takeaway lessons from past failures

#### Learning Review

What does "insignificant point of difference" mean as a reason for new-product failure?

> "Insignificant point of difference" means that a distinctive product feature may have been absent from the new-product.  
> A "point of difference" is necessary for a new product to defeat competitive ones -- through the possession of superior characteristics that deliver unique benefits to the user.

### The New-Product Process

1. New-product strategy development
    * Purpose of stage
        * identify focus, structure, approach, and guidelines for innovation
    * Marketing information and methods used
        * Cross-functional teams
        * Six Sigma
        * Lead users
2. Idea generation
    * Purpose of stage
        * develop concepts for possible products
    * Marketing information and methods used
        * customer and supplier suggestions
        * employee and co-worker suggestions
        * research and development breakthroughs
        * competitive products
        * universities, inventors, and small technology firms
3. Screening and evaluation
    * Purpose of stage
        * separate good product ideas from bad ones inexpensively
    * Marketing information and methods used
        * Internal approach
        * External approach
            * Concept tests
4. Business analysis
    * Purpose of stage
        * identify the product's feature and its marketing strategy, and make financial projections
    * Marketing information and methods used
        * product's key features
        * anticipated marketing mix strategy
        * economic, marketing, production, legal, and profitability analyses
5. Development
    * Purpose of stage
        * create the prototype product, and test it in the laboratory and on consumers
    * Marketing information and methods used
        * laboratory and consumer tests on product prototypes
6. Market testing
    * Purpose of stage
        * test product and marketing strategy in the marketplace on a limited scale
    * Marketing information and methods used
        * standard test markets
        * controlled test markets
        * simulated test markets
        * ...when test markets do not work
7. Commercialization
    * Purpose of stage
        * position and offer product in the marketplace in full-scale and sales
    * Marketing information and methods used
        * perceptual maps
        * product positioning
        * regional roll-outs
    * Slotting fee
    * Failure fee

#### Learning Review

What is the difference between internal and external screening and evaluation approaches used by a firm in the new-product process?

> Internal approach means the firm evaluates the technical feasibility of the proposal and whether the idea meets the objectives defined in the new-product strategy development step.  
> This is all done "in house".
>
> External evaluations of concept tests consist of preliminary testing of the new-product idea (rather than the actual product) with consumers.

What is a test market?

> Market testing is the exposing of actual products to prospective consumers under realistic purchase conditions to see if they will buy.

### The Service Economy

Services

> Intangible activities, benefits, or satisfactions that an organization provides to consumers in exchange for money or something else of value.

Responsible for over 70% of Canada's total economic output, as well as over 75% of its workforce.

#### The Uniqueness of Services

The 4 I's of services

* Intangibility
* Inconsistency
* Inseparability
    * Hard to separate the person delivering the service from the brand itself
        * Bad employee -> poor perception of the service received by the consumer
* Inventory
    * Idle production capacity

#### Service Continuum

$$\text{Service-dominated offering (intangible)} \longleftarrow~\text{Balanced}~\longrightarrow \text{Product-dominated offering (tangible)}$$

#### How Consumers Purchase Services

* Purchasing a Service
    * 3 aspects of a service
        * Search Properties - Determined before purchase (e.g., price)
        * Experience Properties - Experience properties may only be discerned during or after consumption (e.g., restaurant food quality)
        * Credence Properties

Credence properties may be impossible to evaluate (e.g., medical diagnosis).

* Post-purchase Evaluation
    * Gap analysis
        * expectation vs reality
    * SERVQUAL
        * service quality
        * multidimensional research instrument designed to measure service quality by capturing respondents' expectations and perceptions along the five dimensions of service quality
            * Reliability (5 questions)
            * Assurance (4 questions)
            * Tangibles (4 questions)
            * Empathy (5 questions)
            * Responsiveness (4 questions)
        * 22 expectation items and 22 perceptions items

### The Communication Process

* Communication
    * Communication is the process of conveying a message to others and requires a **source**, a **message**, a **channel of communication**, a **receiver**, and the **processes of encoding and decoding**
    * Source - is a sender -- such as a company or salesperson -- who has information to convey
    * Message - is the information to be transmitted, such as the details of a new weight-reduction drink
    * Receivers - are consumers who read, hear, or see the message
* Encoding and Decoding
    * Field of experience
* Feedback
    * Response
* Noise
    * e.g., a misprint in a newspaper ad

### The Promotional Elements

The promotional mix:

* Advertising
    * Mass
    * Payment
        * Fees paid for space or time
    * Strengths
        * Efficient means for reaching large numbers of people
    * Weaknesses
        * High absolute cost
        * Difficult to receive good feedback
* Personal selling
    * Customized
    * Payment
        * Fees paid to salespeople as either salaries or commissions
    * Strengths
        * Immediate feedback
        * Very persuasive
        * Can select audience
        * Can give complex information
    * Weaknesses
        * Extremely expensive per exposure
        * Messages may differ between salespeople
* Public relations
    * Mass
    * Payment
        * No direct payment to media
    * Strengths
        * Often most credible source in the consumer's mind
    * Weaknesses
        * Difficult to get media cooperation
* Sales promotion
    * Mass
    * Payment
        * Wide range of fees paid, depending on promotion selected
    * Strengths
        * Effective at changing behaviour in short run
        * Very flexible
    * Weaknesses
        * Easily abused
        * Can lead to promotion wars
        * Easily duplicated
* Direct marketing
    * Customized
    * Payment
        * Cost of communication through mail, telephone, or computer
    * Strengths
        * Messages can be prepared quickly
        * Facilitates relationship with customer
    * Weaknesses
        * Declining customer response
        * Database management is expensive
